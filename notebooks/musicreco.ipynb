```python
selected_features = [
    "name",
    "artists",
    "danceability",
    "energy",
    "acousticness",
    "instrumentalness",
    "speechiness",
    "valence",
    "tempo",
    "liveness",
    "popularity"
]

songs = df[selected_features].copy()

songs.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
audio_features=["danceability",
    "energy",
    "acousticness",
    "instrumentalness",
    "speechiness",
    "valence",
    "tempo",
    "liveness",
    "popularity"]
```

```python
scaler= StandardScaler()
```

```python
scaled_features= scaler.fit_transform(songs[audio_features])
```

```python
scaled_features.shape
```

```python
scaled_features[:5]
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
model= NearestNeighbors(
    metric="cosine",
    algorithm="brute"
)
```

```python
model.fit(scaled_features)
```

```python
songs[songs["name"]=="Apocalypse"]
```

```python
def recommend(song_name, n_recommendations=10):
    matches= songs[songs["name"].str.lower()==song_name.lower()]
    
    if matches.empty:
        print("song not found")
        return
    
    index=matches.index[0]
    
    distances, indices= model.kneighbors(
        [scaled_features[index]], 
        n_neighbors=n_recommendations+1
    )
    
    print(f"\n recommendation for {song_name}: \n")
    
    for idx in indices[0][1:]:
        print(
            songs.iloc[idx]["name"],
            "--",
            songs.iloc[idx]["artists"]
        )
```

```python
recommend("Apocalypse")
```
